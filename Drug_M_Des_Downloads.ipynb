```python
import numpy as np
import requests
import pandas as pd
data = pd.read_csv(r'C:\Users\venug\Desktop\SInfo.txt',dtype = str,delimiter = '\n')
print(type(data))
SMILES = {}
for i in data['DB_ID']:
    
    url = r'https://www.drugbank.ca/structures/small_molecule_drugs/{}.smiles'.format(i)
    response = requests.get(url)
    print(response.text)
    SMILES[i] = response.text
#SMILES = np.array(SMILES)
#data = list(data)
SMILES = pd.from_dict(SMILES,orient = 'index',columns = ['DB_ID','SMILES'])
pd.merge(data,SMILES,on = 'DB_ID')

print(type(SMILES))
print(data,SMILES)

data = list(zip(data,SMILES))
df = pd.DataFrame(data,columns = ['Accession','SMILES'])

df.to_csv(r'C:\Users\venug\Desktop\ACCESSION_ID_SMILES.csv',index = False)
```
